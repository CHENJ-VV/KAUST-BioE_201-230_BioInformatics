```python
def hammingDistance(string_p, string_q):
    n_p = len(string_p)
    n_q = len(string_q)
    if n_p != n_q:
        print('Two DNA strings do not have the same length')
    hammingDis = 0
    
    for i in range(n_p):
        if string_p[i] != string_q[i]:
            hammingDis += 1
    
    return hammingDis
```
