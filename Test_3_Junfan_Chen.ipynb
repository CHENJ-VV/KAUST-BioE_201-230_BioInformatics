```python
def ComplimentaryDNA(text):
    n = len(text)
    complimentaryStrand = []
    for i in range(n):
        if text[i] == 'A':
            complimentaryStrand.append('T')
        elif text[i] == 'C':
            complimentaryStrand.append('G')
        elif text[i] == 'G':
            complimentaryStrand.append('C')
        elif text[i] == 'T':
            complimentaryStrand.append('A')
    return complimentaryStrand
            
```

```python
def reverseDirection(complimentaryStrand):
    n = len(complimentaryStrand)
    reverseStrand = []
    for i in range(n):
        reverseStrand.append(complimentaryStrand[n - 1 - i])
    reverseStrand = ''.join(reverseStrand)
    return reverseStrand
    
```
