```python
def countPattern(substring, string):
    n_p = len(string)
    k_p = len(substring)
    count_p = 0
    for i in range(n_p - k_p + 1):
        if string[i : i + k_p] == substring:
            count_p += 1
    return count_p

def patternTimes(string, k, L):
    n = len(string)
    count = [0] * (n - L + 1)
    for i in range(n - L + 1):
        interval = string[i : i + L]
        count[i] = countPattern(string[i : i + k], interval)
        # print('i = ', i, 'count = ', count[i])
    return count

def extractPattern(count, t, string, k):
    n = len(string)
    n_l = len(count)
    pattern = set()
    for i in range(n_l):
        # print('i_e = ', i)
        if count[i] >= t:
            pattern.update({string[i : i + k]})
            # print(pattern)
            
    # Codes below are for the last interval check
    count_l = [0] * (L - k + 1)
    interval_l = string[n - L : n]
    for i in range(L - k):
        count_l[i] = countPattern(interval_l[i : i + k], interval_l)
        # print('count_l is as below: ', count_l)
    for i in range(len(count_l)):
        if count_l[i] >= t:
            pattern.update({interval_l[i : i + k]})
            
    pattern = ' '.join(pattern)
    return pattern
```
