```python
def OccurenceOfPattern(stringPattern, Genome):
    occurePosition = []
    n = len(Genome)
    k = len(stringPattern)
    for i in range(n - k + 1):
        if Genome[i : i + k] == stringPattern:
            occurePosition.append(i)
    return occurePosition
```
