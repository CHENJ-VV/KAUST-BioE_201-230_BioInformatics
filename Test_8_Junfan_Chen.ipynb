```python
def hammingDistance(substring_1, substring_2):
    n_1 = len(substring_1)
    n_2 = len(substring_2)
    hammingCount = 0
    if n_1 != n_2:
        print('Two strings do not have the same length!')
        exit()
    for i in range(n_1):
        if substring_1[i] != substring_2[i]:
            hammingCount += 1
    return hammingCount
```

```python
def findLocation(pattern, DNAstring, d):
    k = len(pattern)
    n = len(DNAstring)
    location = []
    for i in range(n - k + 1):
        hammingCompare = hammingDistance(pattern, DNAstring[i : i + k])
        if hammingCompare <= d:
            location.append(i)
    location_output = ' '.join(str(i) for i in location)
    return location_output
```
