```python
def skewFind(string):
    n = len(string)
    skew = [0] * (n + 1)
    skew[0] = 0
    count_C = 0
    count_G = 0
    for i in range(n):
        if string[i] == 'C':
            count_C += 1
        if string[i] == 'G':
            count_G += 1
        # print('count_C: ', count_C)
        # print('count_G: ', count_G)
        skew[i + 1] = count_G - count_C
        # print('minus: ', (count_G - count_C))  
    return skew


def minLocation(skew):
    minimalLocation = []
    n_s = len(skew)
    minValue = min(skew)
    for i in range(n_s):
        if skew[i] == minValue:
            minimalLocation.append(i)
            
    return minimalLocation
```

```python
if __name__ == '__main__':
    string = ''
    skew = skewFind(string)
    minLo = minLocation(skew)
    
    skew_output = ' '.join(str(i) for i in skew)
    minLo_output = ' '.join(str(i) for i in minLo)
    # print('skew is: ', skew_output)
    print('The location(s) of minimal skew are:', minLo_output)
    
```
