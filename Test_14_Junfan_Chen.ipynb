```python
def hammingDistance(pattern1, pattern2, d):
    n_1 = len(pattern1)
    n_2 = len(pattern2)
    hammingCo = 0
    if n_1 != n_2:
        print('Error in pattern selection')
        exit()
    for i in range(n_1):
        if pattern1[i] != pattern2[i]:
            hammingCo += 1
    return hammingCo
```

```python
from functools import reduce
def patternList(pattern):
    k = len(pattern)
    basepair = ['A', 'C', 'G', 'T']
    patternLi = list(reduce(lambda x, y: [pair_1 + pair_2 for pair_1 in x for pair_2 in y], [basepair] * k))
    return patternLi
```

```python
def countList(patternList, pattern, d):
    n = len(patternList)
    count = [9] * n
    for i in range(n):
        count[i] = hammingDistance(patternList[i], pattern, d)
    return count
```

```python
def patternFind(patternList, count, d):
    n = len(count)
    patternCorrect = []
    n_judge = len(patternList)
    if n != n_judge:
        print('Error in dismatch')
    for i in range(n):
        if count[i] <= d:
            patternCorrect.append(patternList[i])
    #patternCorrect = ' '.join(patternCorrect)
    return patternCorrect
```
