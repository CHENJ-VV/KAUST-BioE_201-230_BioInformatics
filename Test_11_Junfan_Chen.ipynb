```python
def countPattern(pattern, Text, k):
    n = len(Text)
    count = 0
    for i in range(n):
        if Text[i : i + k] == pattern:
            count += 1
    return count
```

```python
from functools import reduce
def basePairAssign(k):
    basepair = ['A', 'C', 'G', 'T']
    patternList = list(reduce(lambda x, y: [pair_1 + pair_2 for pair_1 in x for pair_2 in y], [basepair] * k))
    return patternList
```

```python
def countTime(Text, patternList, k):
    n = len(Text)
    n_p = len(patternList)
    countArray = [9] * n_p
    for i in range(n_p):
        countArray[i] = countPattern(patternList[i], Text, k)
    countArray = ' '.join(str(i) for i in countArray)
    return countArray
```
